# Logistic Regression

```python
# can we use cuda?
if torch.cuda.is_available():
    avDev = torch.device("cuda")
else:
    avDev = torch.device("cpu")

print(avDev)
```

```python
torch.manual_seed(123)
```

```python
# making dataset iterable
batch_size = 100
n_iters = 50000
num_epochs = n_iters / (len(trainset) / batch_size)
num_epochs = int(num_epochs)

train_loader = torch.utils.data.DataLoader(dataset=trainset, 
                                           batch_size=batch_size, 
                                           shuffle=True)
 
test_loader = torch.utils.data.DataLoader(dataset=testset, 
                                          batch_size=batch_size, 
                                          shuffle=False)

dataloaders = {
    "train": train_loader,
    "validation": test_loader
}
```

```python
# create model class
class LogisticRegressionModel(nn.Module):
    def __init__(self, input_dim, output_dim, hidden_dim):
        super(LogisticRegressionModel, self).__init__()
        self.linear1 = nn.Linear(input_dim, hidden_dim)
        self.linear2 = nn.Linear(hidden_dim, output_dim)
        self.relu = nn.ReLU()
     
    def forward(self, x):
        out = self.relu(self.linear1(x))
        out = self.linear2(out)
        return out
```

```python
# train model using livelossplot
# This was adapted from a post from Joshy Cyriac on 29 Mar 2020
# GitHub here:
# https://github.com/stared/livelossplot/blob/master/examples/pytorch.ipynb
# retrieved in June 2020.

def train_model(model, criterion, optimizer, num_epochs, cur_epoch = 0):
    min_epoch_loss = np.inf
    
    input_dim = 32*32*3
    liveloss = PlotLosses()
    
    # can we use cuda?
    if torch.cuda.is_available():
        avDev = torch.device("cuda")
    else:
        avDev = torch.device("cpu")
    model.to(avDev)
    
    
    
    for epoch in range(num_epochs):
        logs = {}
        
        for phase in ['train', 'validation']:
            if phase == 'train':
                model.train()
            else:
                model.eval()
                
                
            running_loss = 0.0
            running_corrects = 0

            # for each epoch first train than validation
            for images, labels in dataloaders[phase]:
                images = images.view(-1, input_dim).to(avDev)
                labels = labels.to(avDev)
                
                # Forward pass to get output/logits
                outputs = model(images)
                
                # Calculate Loss: softmax --> cross entropy loss
                loss = criterion(outputs, labels)
                
                if phase == 'train':
                    # Clear gradients w.r.t. parameters
                    optimizer.zero_grad()
                    # Getting gradients w.r.t. parameters
                    loss.backward()
                    # Updating parameters
                    optimizer.step()
                
                
                
                # loss for each batch
                running_loss += loss.detach() * images.size(0)
                
                # correct predictions for each batch
                _, predicted = torch.max(outputs.data, 1)
                running_corrects += (predicted.cpu() == labels.cpu()).sum().float()
                
            epoch_loss = running_loss / len(dataloaders[phase].dataset)
            epoch_acc = running_corrects.float() / len(dataloaders[phase].dataset) * 100.

            prefix = ''
            if phase == 'validation':
                prefix = 'val_'
                
                # new min of epoch loss validation
                if epoch_loss.item() < min_epoch_loss:
                    min_epoch_loss = epoch_loss.item()
                    # save the best model
                    torch.save(model.state_dict(), model_path)

            logs[prefix + 'log loss'] = epoch_loss.item()
            logs[prefix + 'accuracy'] = epoch_acc.item()
            
        liveloss.update(logs)
        liveloss.send()
        
        print('current epoch: ' + str(cur_epoch))
        # checkpoint
        torch.save({
            'epoch': cur_epoch,
            'model_state_dict': model.state_dict(),
            'optimizer_state_dict': optimizer.state_dict(),
            'loss': loss
            }, model_checkpoint_path)
        cur_epoch += 1
        
    return min_epoch_loss
```

```python
from hyperopt import hp, tpe, fmin
space = [hp.uniform('learning_rate', 0.001, 0.005), hp.choice('hidden_dim', [200, 250, 300]), hp.choice('optimizer', ['SGD', 'Adam']), hp.choice('batch_size', [50, 100, 200])]

def f(args):
    torch.manual_seed(123)
    
    learning_rate, hidden_dim, opt, batch_size = args
    input_dim = 32*32*3
    output_dim = 10
    
    # instantiate model class 
    model = LogisticRegressionModel(input_dim, output_dim, hidden_dim)
    
    # instantiate loss class
    criterion = nn.CrossEntropyLoss().to(avDev)
    
    # instantiate optimizer
    if opt == 'SGD':
        optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)
    elif opt == 'Adam':
        optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
        
    train_loader = torch.utils.data.DataLoader(dataset=trainset, batch_size=batch_size, shuffle=True)
    test_loader = torch.utils.data.DataLoader(dataset=testset, batch_size=batch_size, shuffle=False)
    dataloaders = {"train": train_loader, "validation": test_loader}
        
    return train_model(model, criterion, optimizer, num_epochs=100)
    
best = fmin(f, space, algo=tpe.suggest, max_evals=100)

print('tpe: ', best)
```

```python
#loading best model

input_dim = 32*32*3
output_dim = 10
hidden_dim = 300
learning_rate = 0.001

# instantiate model class 
model = LogisticRegressionModel(input_dim, output_dim, hidden_dim)

model.load_state_dict(torch.load(model_path))
```

```python
# can we use cuda?
if torch.cuda.is_available():
    avDev = torch.device("cuda")
else:
    avDev = torch.device("cpu")
model.to(avDev)

# confusion matrix
cm = np.zeros((10,10), dtype=np.int)

with torch.no_grad():     
    # Iterate through test dataset
    for images, labels in test_loader:
        images = images.view(-1, input_dim).to(avDev)

        # Forward pass only to get logits/output
        outputs = model(images)

        # Get predictions from the maximum value
        _, predicted = torch.max(outputs.data, 1)
        
        cm += confusion_matrix(labels.tolist(), predicted.tolist())
```

```python
# This code is for plotting a confusion matrix we already calulated
# input: confusion matrix and class label names
# Source here:
# https://scikit-learn.org/0.18/auto_examples/model_selection/plot_confusion_matrix.html
# retrieved in June 2020.

import itertools
import numpy as np
import matplotlib.pyplot as plt

def plot_confusion_matrix(cm, classes, normalize=False, title='Confusion matrix', cmap=plt.cm.Blues):
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt), horizontalalignment="center", color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
plt.figure(figsize=(10,10))
plot_confusion_matrix(cm, CLASSES)
```

```python
plt.figure(figsize=(10,10))
plot_confusion_matrix(cm, CLASSES, normalize=True)
```

```python
import glob
import os.path
from pathlib import Path
import cv2

src = Path('.')
count = 0
for file in src.rglob('*.mp4'):
    
    print(file)
    vidcap = cv2.VideoCapture(str(file))
    success,image = vidcap.read()

    success = True
    while success:
        success,image = vidcap.read()
        if(count%10 == 0):
            try:
                cv2.imwrite("img/frame%d.jpg" % count, image)
            except:
                pass
        if cv2.waitKey(10) == 27:
            break
        count += 1
```
